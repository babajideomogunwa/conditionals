```python
x = np.linspace(-2*np.pi, 2*np.pi)
y1 = np.sin(x)
y2 = np.cos(x)

fig, axes = plt.subplots(1,2)
axes[0].plot(x, y1, color = 'r', linewidth=5)
axes[1].plot(x, y1, color='#ffaa00', linewidth=0.5, linestyle='--')
axes[1].plot(x, y2, color='green', linestyle='-.')

#set the current axis to the first subplot
fig.sca(axes[0])
#set x and y ticks
plt.xticks(ticks=[-np.pi, -np.pi/2, 0, np.pi/2, np.pi], labels=["Zero", "One", "Two", "Three", "Four"])
plt.yticks(ticks=[-1, 0, 1], labels=["Low", "Medium", "High"])

#set x and y limits
plt.xlim(-np.pi, np.pi)
plt.ylim(-1, 1)

plt.title("Custom Tick Labels")
plt.xlabel("Index")
plt.ylabel("Value")

#set the current axis to the second subplot
fig.sca(axes[1])
#set x and y ticks and labels
plt.xticks(ticks=[-np.pi, -np.pi/2, 0, np.pi/2, np.pi], labels=["Zero", "One", "Two", "Three", "Four"])
plt.yticks(ticks=[-1, 0, 1], labels=["Low", "Medium", "High"])

#set x and y limits
plt.xlim(-np.pi, np.pi)
plt.ylim(-1, 1)

plt.title("Custom Tick Labels")
plt.xlabel("Index")
plt.ylabel("Value")
plt.grid(True)
plt.show()

plt.tight_layout();
```
