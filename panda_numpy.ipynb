```python
# Import the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# read in the files
file1 = pd.read_csv('task2a.csv')
file2 = pd.read_csv('task2b.csv')
```

```python
# Observations: The files have equal number of rows, 6468, and 2 columns: (entity and year) are similar in both files.
# Both files can be merged based on the common columns with inner join

# merging the two files
all_data = pd.merge(file1,file2, on=['Entity', 'Year'], how= 'inner')
all_data
```

```python
all_data.shape
```

```python
# Checking the data types of the columns
all_data.info()
```

```python
# Getting summary statistics
all_data.describe()
```

- The statistical summary shows that anxiety disorders has the highest percentage (8.97%) while eating disorders has the minimum percentage (0.074%)

```python
# all_data.dropna
# all_data.drop(columns='Entity')
```

```python
# Checking for correlation between the different columns

# Select only numeric columns from the DataFrame
numeric_data = all_data.select_dtypes(include=['number'])

# Compute correlation on numeric data only
correlation = numeric_data.corr()

# Display the correlation matrix
print(correlation)
```

```python
# Plotting the correlation among columns
plt.figure(figsize=(15, 7))
sns.heatmap(correlation, annot=True, vmin=-1, vmax=1, fmt=".2f", cmap="Spectral")
plt.show()
```

### Interesting Insights
- Drug use abuse is positively correlated with anxiety disorders, eating disorders and schizophrenia
- Eating, anxiety and bipolar disorders are strongly positively correlated
- Alcohol use disorders is negatively correlated with anxiety disorders

```python
# Checking the country with the highest drug use abuse

avg_drug_use = all_data.groupby(['Entity']).mean()[['Drug use disorders (%)']].reset_index()
avg_drug_use
```

```python
avg_drug_eating_dis = all_data.groupby(['Entity']).mean()[['Drug use disorders (%)', 'Eating disorders (%)']].reset_index()
avg_drug_eating_dis
```

```python
# Sorting in descending order using sort_values function
avg_drug_use = avg_drug_use.sort_values(ascending=False, by='Drug use disorders (%)')
avg_drug_use
```

- We can see that the USA has the highest rate of drug use disorders while Bosnia and Herzegovina has the lowest

```python
avg_drug_eating_dis = avg_drug_eating_dis.sort_values(ascending=False, by='Eating disorders (%)')
avg_drug_eating_dis
```

```python
# Checking the relationship between drug use and anxiety disorders
avg_drug_anxiety_dis = all_data.groupby(['Entity']).mean()[['Drug use disorders (%)', 'Anxiety disorders (%)']].reset_index()
#avg_drug_anxiety_dis = avg_drug_anxiety_dis.sort_values(ascending=False, by='Anxiety disorders (%)')
avg_drug_anxiety_dis
```

```python
sns.relplot(
    data=avg_drug_anxiety_dis,
    x='Drug use disorders (%)', y='Anxiety disorders (%)'
)
```

- there are outliers between anxiety disorders and drug use disorders

```python
# We can look at the relationship between eating disorders and bipolar disorders
eating_bipolar = all_data.groupby(['Entity']).mean()[['Eating disorders (%)', 'Bipolar disorder (%)']].reset_index()
eating_bipolar
```

```python
sns.relplot(
    data=eating_bipolar,
    x='Eating disorders (%)', y='Bipolar disorder (%)'
)
```

```python
# We can look at the relationship between eating disorders and bipolar disorders
alcohol_anxiety= all_data.groupby(['Entity']).mean()[['Alcohol use disorders (%)', 'Anxiety disorders (%)']].reset_index()
alcohol_anxiety
```

```python
# Plotting the relationship between Alcohol use disorders amd Anxiety disorders
sns.relplot(
    data=alcohol_anxiety,
    x='Anxiety disorders (%)', y='Alcohol use disorders (%)'
)
```

- The above graph shows they are not positively correlated

- From the correlation analysis, as more drug use disorders increase, the more the anxiety disorders increase as well, i.e they are positively correlated.
- However, before coming to conclusion, it will be checked if there are outliers that can skew the data.

```python
# Visualising the data
# Checking whether the data are normally distributed
# Distplot has been deprecated, histplot is used instead and specifying density in stat

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.histplot(avg_drug_anxiety_dis['Drug use disorders (%)'], kde = True, stat= 'density', kde_kws = dict(cut=3), alpha=.4, color='skyblue')
plt.subplot(1,2,2)
sns.histplot(avg_drug_anxiety_dis['Anxiety disorders (%)'], kde=True, stat='density', kde_kws= dict(cut=3), alpha=.4, color='salmon')
plt.tight_layout()
plt.show()
```
